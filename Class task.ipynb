```python
import numpy as np
```

```python
iris = np.genfromtxt(r"C:\Users\Admin\Downloads\Iris_clean_-_Copy.csv",dtype = 'float32', delimiter = ',')
iris
```

```python
sepal_len = iris[1:,0] # sepal length
sepal_len
```

```python
sepal_len = np.where(sepal_len>=50,2,sepal_len)
sepal_len
```

```python
sepal_len = np.where(sepal_len==-1,np.nan,sepal_len)
sepal_len
```

```python
med = np.nanmedian(sepal_len)
med
```

```python
sepal_len = np.where(np.isnan(sepal_len),med,sepal_len)
sepal_len
```

```python
sepal_l = sepal_len.dtype = 'float32'
sepal_l
```

sepal_width

```python
sepal_wid = iris[1:,1] # sepal width
sepal_wid
```

```python
sepal_wid = np.where(sepal_wid>=40,3,sepal_wid)
sepal_wid
```

```python
sepal_wid = np.where(sepal_wid==-1,np.nan,sepal_wid)
sepal_wid
```

```python
med1 = np.nanmedian(sepal_wid)
med1
```

```python
sepal_wid = np.where(np.isnan(sepal_wid),med1,sepal_wid)
sepal_wid
```

```python
petal_len = np.where(petal_len>=10,np.nan,petal_len)
petal_len
```

```python
med2 = np.nanmedian(petal_len)
med2
```

```python
petal_len = np.where(np.isnan(petal_len),med2,petal_len)
petal_len
```

petal_width
